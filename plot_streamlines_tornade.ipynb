```python
import matplotlib.pyplot as plt
import numpy as np
```

Écoulement d'une tornade
\begin{align}
V_r = -\frac{C_1}{r} \\
V_\theta = -\frac{C_2}{r} 
\end{align}

On peut jouer sur $C_1$ et $C_2$ pour obtenir une tornade sympathique, $C_2$ contrôle l'enroulement des lignes de courant

```python
C1 = 1.0
C2 = 3.0
```

Les lignes de courant sont les courbes 
\begin{align}
\theta = \frac{C_2}{C_1} \ln{r} + \text{cst}
\end{align}

```python
nb_ligne = 20
cst = np.linspace(0.,2*np.pi,nb_ligne)
```

```python
r = np.linspace(0.000001,5.,100)
theta = C2/C1 * np.log(r) 
for i in range(nb_ligne):
    x = r*np.cos(theta+cst[i])
    y = r*np.sin(theta+cst[i])
    plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```
